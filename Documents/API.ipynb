```python
import requests
import pandas as pd

# Define the base URL for the API
base_url = "https://clinicaltrials.gov/api/v2/studies"

# Set up the query parameters
params = {
    "query.cond": "Diabetes",
    "pageSize": 10,
    "format": "json",
    "countTotal": "true"
}

# Send the GET request to the API
response = requests.get(base_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    studies = data.get("studies", [])
    
    # Extract relevant information from the studies
    study_data = []
    for study in studies:
        study_info = {
            "NCTId": study["protocolSection"]["identificationModule"]["nctId"],
            "Title": study["protocolSection"]["identificationModule"]["officialTitle"],
            "Status": study["protocolSection"]["statusModule"]["overallStatus"],
            #"Phase": study["protocolSection"]["designModule"]["phase"],
            #"Enrollment": study["protocolSection"]["enrollmentModule"]["enrollmentCount"],
            #"Completion Date": study["protocolSection"]["statusModule"]["completionDate"]
        }
        study_data.append(study_info)
    
    # Convert the extracted data into a DataFrame
    df = pd.DataFrame(study_data)
    print(df)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
import requests
import json

# Define the base URL for the API
base_url = "https://clinicaltrials.gov/api/v2/studies"
params = {
    "query.cond": "Diabetes",
    "pageSize": 10,  # just get two studies to inspect
    "format": "json"
}

response = requests.get(base_url, params=params)

if response.status_code == 200:
    data = response.json()
    studies = data.get("studies", [])
    
    # Filter studies with results
    studies_with_results = [s for s in studies if s.get("hasResults")]

    for i, study in enumerate(studies_with_results):
        print(f"--- Study {i+1} ---")
        print(json.dumps(study, indent=4))
else:
    print(f"Failed to retrieve data: {response.status_code}")


```
